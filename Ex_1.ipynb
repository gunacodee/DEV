```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Sample employee data
data = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'IT'],
    'Experience (Years)': [3, 5, 2, 7, 4],
    'Performance Score': [85, 90, 78, 88, 92]
}

# Step 3: Create DataFrame
df = pd.DataFrame(data)
print("DataFrame:")
print(df)

# Step 4: Data Analysis
print("\nData Analysis:")
print("Average Experience:", df['Experience (Years)'].mean())
print("Highest Performance Score:", df['Performance Score'].max())
print("Lowest Performance Score:", df['Performance Score'].min())

# Step 5: Data Visualization
plt.figure(figsize=(8, 4))
plt.bar(df['Employee'], df['Performance Score'], color='skyblue')
plt.xlabel('Employee')
plt.ylabel('Performance Score')
plt.title('Employee Performance Comparison')
plt.grid(True)
plt.tight_layout()
plt.show()
```
