```python
import numpy as np

# Create a NumPy array representing sales of products in different regions
sales = np.array([[150, 200, 250], [300, 400, 500]])
print("Array:\n", sales)
print("Type:", type(sales))
print("Dimensions:", sales.ndim)
print("Shape:", sales.shape)
print("Size:", sales.size)
print("Data type:", sales.dtype)

# Slicing
print("Sales in Region 2:", sales[1, :])

# Column-wise access
print("Sales of Product 2 in all regions:", sales[:, 1])
```

```python
import pandas as pd
import numpy as np

# Create a DataFrame for student marks
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [88, 75, 95]
}

df = pd.DataFrame(data)
print("Student Marks DataFrame:\n", df)

# Create DataFrame from dictionary of arrays
scores_dict = {101: ['Pass', 'Fail'], 102: ['Pass', 'Pass']}
print("\nFrom dictionary:\n", pd.DataFrame(scores_dict))

# Series to DataFrame
cities_series = pd.Series({"USA": "Washington", "France": "Paris", "India": "New Delhi"})
print("\nFrom Series:\n", pd.DataFrame(cities_series))

# Display shape
print("\nShape of Student DataFrame:", df.shape)
print("Number of Rows:", len(df.index))
```

```python
import matplotlib.pyplot as plt

# Daily closing stock prices
days = [1, 2, 3, 4, 5]
apple = [150, 152, 149, 155, 158]
google = [2800, 2825, 2790, 2850, 2900]

# Basic line plot
plt.figure(figsize=(8, 4))
plt.plot(days, apple, marker='o', label='Apple')
plt.plot(days, google, marker='s', label='Google')
plt.xlabel("Day")
plt.ylabel("Stock Price")
plt.title("Stock Prices Over 5 Days")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Subplots for multiple trends
sales = [20, 30, 40, 50]
revenue = [1000, 1500, 2000, 2500]
profit = [300, 450, 600, 750]

fig = plt.figure(figsize=(10, 10))
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 2, 3)

ax1.plot(sales, 'bo--')
ax1.set_title("Sales")
ax2.plot(revenue, 'g*-')
ax2.set_title("Revenue")
ax3.plot(profit, 'r^-')
ax3.set_title("Profit")

plt.tight_layout()
plt.show()
```
